```python
import math
import numpy as np
from matplotlib import pyplot as plt
def func_2_29(eps:float,N:int):
    def f(x:float):
        return np.sin(x) ** 2 + np.cos(2 * x + 1)
    def f_diff(x:float):
        return np.sin(2 * x) + 2 * np.sin(2*x + 1)
    X = []
    Y = []
    for _ in np.arange(-3,2,0.01):
        X.append(_)
        Y.append(f(_))
    plt.plot(X,Y)
    def find_root(a0,b0,eps):
        x0 = b0
        x1 = x0 + f(x0) / f_diff(x0)
        while abs(x1 - x0) >= eps * abs(a0 - b0):
            x0 = x1
            x1 = x1 + f(x1) / f_diff(x1)
        return x1
    def calc_int(x1, x2, N):
        delta_x = (x2 - x1) / N
        integral = 0
        for _ in range(N):
            integral += delta_x * (f(x2 + _ * delta_x) + f(x1 + _ * delta_x)) / 2
        return integral
    r1 = find_root(-2.5,-1.5,eps)
    r2 = find_root(0,1,eps)
    integral = calc_int(r1, r2, N)
    return r1,r2,integral



def func_order_of_acuracy(n1,n2,eps):  #Функция определения порядка точности   (n2 > n1) - количество разбиений
    string1 = func_2_29(eps,n1)
    string2 = func_2_29(eps,n2)
    integral1 = string1[2]
    integral2 = string2[2]
    r1 = string1[0]
    r2 = string1[1]
    delta_x1 = (r2 - r1) / n1
    delta_x2 = (r2 - r1) / n2
    A = []
    B = []
    for _ in range(n1 + 1):
        x = r1 + _ * delta_x1
        inaccuracy1 = (2 * np.cos(2*x) - 2 * np.cos(2*x + 1)) * ((string1[1] - string1[0]) ** 3) / (12 * n1 ** 2)
        A.append(inaccuracy1)
    for _ in range(n2 + 1):
        x = r1 + _ * delta_x2
        inaccuracy2 = (2 * np.cos(2*x) - 2 * np.cos(2*x + 1)) * ((string1[1] - string1[0]) ** 3) / (12 * n2 ** 2)
        B.append(inaccuracy2)
    inacuracy1 = max(A)
    inacuracy2 = max(B)
    fraction = delta_x1 / delta_x2
    integral_fraction = inaccuracy1 / inaccuracy2
    order_of_accuracy = math.log(integral_fraction,fraction)
    return order_of_accuracy
```
